## Zapiski

Implementacijo, opis in vrednotenje metod za nadzorovanjo učenje vsebujejo knjižnice `sklearn` ali `Orange`.

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor

from sklearn.metrics import mean_squared_error
```

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
movies  = pd.read_csv('ml-latest-small/movies.csv')
```

##### m = vsi filmi s 100 ali več ogledov
##### n = vsi uporabniki, ki so si ogledali 100 ali več filmov

```python
# grupiramo po movieId
rating_groups = ratings.groupby('movieId', as_index = False).agg({'timestamp': np.size})
# id-ji filmov, ki imajo 100 ali več ogledov
movie_ids = (rating_groups[rating_groups["timestamp"] >= 100])["movieId"]
# grupiramo po userId
rating_groups = ratings.groupby('userId', as_index = False).agg({'timestamp': np.size})
# uporabniki, ki so si ogledali 100 ali več filmov
user_ids = (rating_groups[rating_groups["timestamp"] >= 100])["userId"]
# grupiramo po userId, da bi lahko dobili vse filme, ki so si jih vsi uporabniki ogledali
rating_groups = ratings.groupby('userId', as_index = False)
```

```python
# ustvarimo nov dataframe, ki kot stolpce uporabi movieId in vsebino seznama user_id (id-ji uporabnikov)
new_df = pd.DataFrame(columns=['movieId'] + list(user_ids))
```

```python
# v data frame dodamo vse id-je filmov
for id in movie_ids:
    new_df = new_df.append({"movieId": id}, ignore_index=True)
```

```python
# ustavimo vrednosti v dataframe
for a in rating_groups.groups:
    temp = rating_groups.get_group(a)["userId"]
    if len(temp) < 100:
        continue
    user_id = rating_groups.get_group(a)["userId"]
    user_id = user_id.iloc[0]
    movie_id = rating_groups.get_group(a)["movieId"]
    rating = rating_groups.get_group(a)["rating"]
    for id, r in zip(movie_id, rating):
        if (id == new_df["movieId"]).any():
            (new_df[new_df["movieId"] == id])
            new_df.at[id, user_id] = r
# zamenjamo neznane vrednosti (Nan) z 0
new_df = new_df.fillna(0);
```

```python
# spremenimo podatkovni tip movieId stolpca v int
new_df["movieId"] = new_df['movieId'].astype(int)
```

```python
matrix = new_df.as_matrix()

linear_regression = LinearRegression()
ridge = Ridge()
lasso = Lasso()
knn_5 =  KNeighborsRegressor(15)
knn_10 =  KNeighborsRegressor(15)
knn_15 =  KNeighborsRegressor(15)
svm = SVR(kernel='rbf', C=1e3, gamma=0.1)

models = [linear_regression, ridge, lasso, knn_5, knn_10, svm, knn_15]
names = ["Linearna regresija", "Ridge", "Lasso", "KNN 5", "KNN 10", "SVM", "KNN 15"]
for m in range(len(names)): 
    total_mse = 0.0
    for column in range(1, matrix.shape[1]):
        current_mse = 0.0
        for i in range(4):
            # nakljucno premesano matriko
            np.random.shuffle(matrix)
            # locimo ucne atribute in ciljni atribut
            y = matrix[:, column]
            x1 = matrix[:, 0:column]
            x2 = matrix[:, column + 1:]
            x = np.concatenate((x1, x2), 1)

            # Razdelimo na ucno in testno mnozico
            y_train = y[:int(len(y) * 0.75)]
            y_test  = y[int(len(y) * 0.75):]

            x_train = x[:int(len(x) * 0.75), :]
            x_test  = x[int(len(x) * 0.75):, :]

            # Ucenje modela
            model = models[m]
            model.fit(x_train, y_train)
            # Napoved vrednosti za podatke
            hx = model.predict(x_test)
            # Evalvacija modela
            mse      = mean_squared_error(hx, y_test)

            current_mse += mse
    
        total_mse += (current_mse / 4)

    total_mse /= matrix.shape[1]
    print(names[m], total_mse)
```

#### KNN15 naredi najmanjšo napako. Če korenimo napako (RMSE) dobimo 1.2, kar pomeni, da model v povprečju zgreši za eno oceno. Napovedi niso zelo slabe in tudi niso zelo dobre.
